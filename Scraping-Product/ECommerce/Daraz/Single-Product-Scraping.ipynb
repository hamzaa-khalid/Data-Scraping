## Importing Libraries

```python
# Importing Libraries
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import time
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import re
```

# Main Script

```python
# Driver Settings
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--incognito")
# chrome_options.add_argument("--headless")
driver = webdriver.Chrome('C:\chromedriver_win32\chromedriver.exe',chrome_options=chrome_options)

def darazScript(lst):
    # Creating None Variables
    imageLink = title = price = rating = totalRating = description = delOptions = waroptions = None
    rewiewsFirstPage = []
    listRat = []
    df = pd.DataFrame(columns=['ImageLink', 'Title', 'Price', 'Rating', 'TotalRatings', 'Description', 'Delivery Options', 'Warranty Options', '1stPageReviews', 'Rating Count'])
    for i in range(len(lst)):
        driver.get(lst[i])
        time.sleep(2)
        
        try:
            itemGallery = driver.find_element_by_xpath("//div[@class='item-gallery']")
            itemGallery2 = itemGallery.find_element_by_tag_name('img')
            imageLink = itemGallery2.get_attribute('src')

            titleWT = driver.find_element_by_xpath("//div[@class='pdp-product-title']")
            title = titleWT.text

            priceWT = driver.find_element_by_xpath("//div[@class='pdp-mod-product-price']//span")
            price = priceWT.text

            ratingWT = driver.find_element_by_xpath("//div[@class='score']")
            rating = ratingWT.text

            totalRatingWT = driver.find_element_by_xpath("//div[@class='count']")
            totalRating = totalRatingWT.text

            revWT = driver.find_elements_by_xpath("//div[@class='item-content']//div[@class='content']")
            for i in revWT:
                rewiewsFirstPage.append(i.text)


            try:
                btn = driver.find_element_by_xpath("//div[@class='expand-button expand-cursor']")
                btn.click()

                desc = driver.find_element_by_xpath("//div[@class='pdp-product-detail']")
                description = desc.text

            except:
                desc = driver.find_element_by_xpath("//div[@class='pdp-product-detail']")
                description = desc.text

            delOpt = driver.find_element_by_xpath("//div[@class='delivery__content']")
            delOptions = delOpt.text


            warOpt = driver.find_element_by_xpath("//div[@class='warranty__options']")
            waroptions = warOpt.text


            opt = driver.find_elements_by_xpath("//span[@class='percent']")
            count = 5
            for i in opt:
                listRat.append(str(count)+' Stars : '+ str(i.text))
                count-=1
            mainList = []
            mainList = [imageLink, title, price, rating, totalRating, description, delOptions, waroptions, rewiewsFirstPage, listRat]
            df_length = len(df)
            df.loc[df_length] = mainList
        except:
            pass
    return df
```

```python
lst = []
f = open("daraz.txt", "r")
data = f.readlines()
for i in data:
    lst.append(i)

# ---------------------------
newdf = darazScript(lst)
newdf
```
