```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
import time
import numpy as np
import pandas as pd
import re
```

### Main Script 1

```python
def categoryAmazon(lsst):
    # Driver Settings
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument("--incognito")
    # chrome_options.add_argument("--headless")
    driver = webdriver.Chrome(ChromeDriverManager().install(),chrome_options=chrome_options)
    df = pd.DataFrame(columns=['ProductLink', 'ImageLink', 'Title', 'TotalRating', 'Rating'])
    for j in range(len(lsst)):
        driver.get(lsst[j])
        paginationlst = driver.find_element_by_xpath("//ul[@class='a-pagination']")
        lastPage = paginationlst.text.split('\n')[-2]

        for i in range(1,int(lastPage)+1):
            c = lsst[0].replace('&dc&qid=','&page={}&qid='.format(i))
            driver.get(c)
            productLink = imageLink = title = totalRating = ratings = None

            #Scraping
            prods = driver.find_elements_by_xpath("//div[@class='a-section a-spacing-medium']")
            pp = driver.find_elements_by_xpath("//div[@class='a-row a-size-small']")
            rr = driver.find_elements_by_xpath("//div[@class='a-row a-size-small']//span[1]")
            time.sleep(2)
            lst = []
            lst2 = []
            for k in rr:
                lst.append(k.get_attribute('aria-label'))
            lst2 = list(filter(lambda a: a != None, lst))

            time.sleep(2)
            for prod,i,j in zip(prods, pp, lst2):
                p1 = prod.find_element_by_tag_name("a")
                productLink = p1.get_attribute('href')
                p2 = prod.find_element_by_tag_name("img")
                imageLink = p2.get_attribute('src')
                p3 = prod.find_element_by_tag_name("h2")
                title = p3.text
                totalRating = i.text
                ratings = j

                mainList = []
                mainList = [productLink, imageLink, title, totalRating, ratings]
                df_length = len(df)
                df.loc[df_length] = mainList
        return df
```

```python
lst = ['https://www.amazon.com/s?i=electronics-intl-ship&bbn=16225009011&rh=n%3A2811119011%2Cn%3A7072561011&dc&qid=1623656301&rnid=2811119011&ref=sr_nr_n_1']
newdf = categoryAmazon(lst)
```
