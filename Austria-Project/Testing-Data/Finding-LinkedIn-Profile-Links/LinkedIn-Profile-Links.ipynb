```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import time
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
lst = list(md.MD)
```

```python
len(lst)
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--incognito")
driver = webdriver.Chrome('C:\chromedriver_win32\chromedriver.exe',chrome_options=chrome_options)
driver.implicitly_wait(2)

k = 'Dipl.-Ing. (FH) Miroslav Nikolovski'
```

```python
find = []
for k in lst:
    links = []
    try:
        i = k+' linkedin austria'
        driver.get('https://www.google.com/search?q='+i+'&oq=google&aqs=chrome..69i57.834j0j4&sourceid=&ie=UTF-8')
        driver.implicitly_wait(3)
        time.sleep(2)
        #--------------------------------------------------------
        soup_a = BeautifulSoup(driver.page_source, 'lxml')
        xx = soup_a.find("div", {"id": "search"})

        #--------------------------------------------------------
        for link in xx.find_all('a'):
            # link.get('href') gets the href/url out of the a element
            links.append(link.get('href'))

        #--------------------------------------------------------
        filter_object = list(filter(lambda a: 'https://at.linkedin.com/in/' in a, links))
        if filter_object != None:
            find.append(filter_object[0])
        else:
            find.append(None)
    except:
        find.append('Not Found')
```

```python
dt = pd.DataFrame(columns=['Links LinkedIn'])
for i in range(len(find)):
    dt = dt.append({'Links LinkedIn': find[i]}, ignore_index=True)
```
